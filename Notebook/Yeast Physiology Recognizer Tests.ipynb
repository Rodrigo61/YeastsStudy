```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

```python
ypds = pd.read_csv('../datasetsYPDS.csv')
ypds = ypds.reindex(np.random.permutation(ypds.index))
ypds
```

```python
train = ypds[:30]
test = ypds[30:]
trainClasses = (train.T)[:1].T
trainData = (train.T)[1:].T
testClasses = (test.T)[:1].T
testData = (test.T)[1:].T
```

```python
knn = KNeighborsClassifier(1, algorithm='ball_tree', weights='uniform')
knn.fit(trainData, trainClasses['00class'])
```

```python
results = knn.predict(testData)
knn.score(testData, testClasses)
```

```python
rf = RandomForestClassifier(n_estimators=50, max_depth=5)
rf.fit(trainData, trainClasses['00class'])
```

```python
rf.score(testData, testClasses)
```

```python
gbc = GradientBoostingClassifier(learning_rate=0.05, n_estimators=200)
gbc.fit(trainData, trainClasses['00class'])
```

```python
results = gbc.predict(testData)
gbc.score(testData, testClasses)
```
